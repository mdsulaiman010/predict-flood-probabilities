### 2. EDA

```python
for feature in df_train.drop(['id','FloodProbability'],axis=1).columns:     # exclude id as its irrelevant, FloodProbability will be viewed separately
    min_val = min(df_train[feature]); max_val = max(df_train[feature])
    levels = np.arange(min_val,max_val+1, 1)
    
    plt.figure(figsize=(5,3))
    plt.bar(levels,df_train[feature].value_counts().sort_index(), color='teal')
    plt.xticks(levels)
    plt.title(f'Distribution of {feature}')
    plt.show()

plt.plot(df_train['FloodProbability'], color='red')
plt.hlines(0.5, color='black', linestyles='--', xmin=0, xmax=df_train.shape[0])
plt.title('Traceplot of Flood Probability')
plt.yticks(np.arange(0,1,0.1))
plt.show()
```

All features present reasonable normality in their respective distributions, with minimal right skew. This is also justified by the mean and median being rather close to each other in value. (makes logistic regression and Gaussian Naive Bayes more feasible for model testing)

Flood probability seem to be roughly around 0.5 based on traceplot, suggests that floods generally occur with the flip of a coin.

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
rescaled_data = ss.fit_transform(df_train.drop(['id','FloodProbability'],axis=1))
rescaled_df_train = pd.DataFrame(rescaled_data, columns=df_train.drop(['id','FloodProbability'],axis=1).columns)
```

```python
from scipy import stats

conclusions = []
pvalues = []

for feature in rescaled_df_train.columns:
    kstest_results = stats.ks_1samp(rescaled_df_train[feature], stats.norm.cdf)
    
    if kstest_results.pvalue > 0.05:
        conclusions.append('Normal')
    else:
        conclusions.append('Not Normal')
    pvalues.append(kstest_results.pvalue)
    

pd.DataFrame({'Feature': rescaled_df_train.columns,
              'p-value': pvalues,
              'Conclusion': conclusions})
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr_fs = DecisionTreeRegressor(random_state=42)
dtr_fs.fit(df_train.drop(['id','FloodProbability'],axis=1), df_train['FloodProbability'])
```

```python
importance_df = pd.DataFrame({'Feature':df_train.drop(['id','FloodProbability'],axis=1).columns,
                              'Feature Importance': dtr_fs.feature_importances_}).sort_values('Feature Importance')

plt.figure(figsize=(12,5))
bars = plt.barh(importance_df['Feature'],importance_df['Feature Importance'], color='teal')
# bars = plt.barh(df_train.drop(['id','FloodProbability'],axis=1).columns ,dtc_fs.feature_importances_, color='teal')
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.3f}', 
             va='center', ha='left')

plt.title('Importance of Feature Variables using DTR')
plt.xlabel('importance')
plt.show()
```

All features have roughly similar importance as each other. Consider alternative feature selection process to further sift of potentially irrelevant features.

```python
from sklearn.feature_selection import mutual_info_regression
info_gain = mutual_info_regression(df_train.drop(['id','FloodProbability'],axis=1), df_train['FloodProbability'])
```

```python
info_df = pd.DataFrame({'Feature':df_train.drop(['id','FloodProbability'],axis=1).columns,
                        'Information Gain': info_gain}).sort_values('Information Gain')

plt.figure(figsize=(12,5))
bars = plt.barh(info_df['Feature'],info_df['Information Gain'], color='teal')
# bars = plt.barh(df_train.drop(['id','FloodProbability'],axis=1).columns ,dtc_fs.feature_importances_, color='teal')
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.3f}', 
             va='center', ha='left')

plt.title('Information Gain from Dataset Feature Variables')
plt.xlabel('Information Gain')
plt.show()
```

Information gained from the features are roughly similar to each other also. Decent evidence to refrain from conducting further dimension reduction techniques. Wrapper-based methods were avoided with interest of time processing findings.

### 3. Partitioning Data

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(df_train.drop(['id','FloodProbability'], axis=1), df_train['FloodProbability'],
                                                      train_size=0.8, random_state=42)
```

```python
print(f'Training observation count: {len(X_train)} (80%)')
print(f'Validation observation count: {len(X_valid)} (20%)')
```

### 4. Model Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_valid)

print(f'MSE: {mean_squared_error(y_valid, y_pred)}')
print(f'R2 Score: {r2_score(y_valid, y_pred)}')
```

```python
import xgboost as xgb
```

```python
dmat_train = xgb.DMatrix(X_train, y_train)
dmat_valid = xgb.DMatrix(X_valid, y_valid)
```

```python
param_dict = {'objective': 'reg:squarederror'}

xgb_reg = xgb.train(param_dict, dtrain=dmat_train, num_boost_round=100, early_stopping_rounds=20, evals=[(dmat_train, 'train')])
y_pred = xgb_reg.predict(dmat_valid)

print(f'MSE: {mean_squared_error(y_valid, y_pred)}')
print(f'R2 Score: {r2_score(y_valid, y_pred)}')
```

```python
from sklearn.ensemble import RandomForestRegressor
rfc_reg = RandomForestRegressor(random_state=42)
rfc_reg.fit(X_train, y_train)
y_pred = rfc_reg.predict(X_valid)

print(f'MSE: {mean_squared_error(y_valid, y_pred)}')
print(f'R2 Score: {r2_score(y_valid, y_pred)}')
```
